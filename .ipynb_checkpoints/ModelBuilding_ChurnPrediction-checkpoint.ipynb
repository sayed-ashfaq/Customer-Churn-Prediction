```python
# importing core libraries
import pandas as pd
import numpy as np

# import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# import libraries for building Model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
```

```python
# importing dataset

data = pd.read_csv("Processed_Churn.csv")
data.head()
```

```python
data = data.drop(labels="Unnamed: 0", axis= 1)
```

The data is heavily imbalanced - With this kind of data most of models give poor results
- We can do upsampling or Downsampling to solve this problem
- Since larger the data better the model - We can do UpSample by creating synthetic data point using SMOTE.

```python
x = data.drop("Churn", axis= 1)
y = data['Churn']
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts().plot(kind= "barh", figsize=(10,3))
plt.xlabel("Count", labelpad= 10)
plt.ylabel("Target Variable", labelpad= 14)
plt.title("Count of Tgt Variable per category", y= 1.02)
plt.show()
```

The data is heavily imbalanced - With this kind of data most of models give poor results
- We can do upsampling or Downsampling to solve this problem
- Since larger the data better the model - We can do UpSample by creating synthetic data point using SMOTE.

```python
sm = SMOTE(random_state= 42)
x_resampled, y_resampled = sm.fit_resample(x, y)
y_resampled.value_counts()
```

```python
# split the data into train and test

x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size= 0.2)

print('X-Train', x_train.shape, "X-test", x_test.shape, "Y-Train", y_train.shape, "Y-test", y_test.shape)
```

```python
x_train
```

### Model Training

```python
models = {
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    "XGBClassifier":XGBClassifier(),
    "RandomForestClassifier": RandomForestClassifier()
}
```

```python
results = {}

for model_name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)


    acc = accuracy_score(y_test, y_pred)
    results[model]= acc

    print(f"Model: {model_name}")
    print(classification_report(y_test, y_pred, labels=[0,1]))
    print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
    print("-" * 50)

# Print final accuracy comparison
for model, score in results.items():
    print(f"{model.__class__.__name__}: Accuracy = {score:.4f}")
```

### Not Satisfied with results
- Since the results were bad giving score that is less than 90% is not making me happy
- We do have multiple technique to solve this like
    - Feature Engineering
    - Experimenting with Sampling method

```python
sen = SMOTEENN()
x_resampled_1, y_resampled_1 = sen.fit_resample(x, y)
```

```python
y_resampled_1.value_counts().plot(kind= "barh", figsize=(10,3))
plt.xlabel("Count", labelpad= 10)
plt.ylabel("Target Variable", labelpad= 14)
plt.title("Count of Tgt Variable per category", y= 1.02)
plt.show()
```

```python
xr_train, xr_test, yr_train, yr_test = train_test_split(x_resampled_1, y_resampled_1, test_size= 0.3)
```

```python
results_ENN = {}

for model_name, model in models.items():
    model.fit(xr_train, yr_train)
    yr_pred = model.predict(xr_test)


    acc = accuracy_score(y_test, y_pred)
    results_ENN[model]= acc

    print(f"Model: {model_name}")
    print(classification_report(yr_test, yr_pred, labels=[0,1]))
    print("Confusion Matrix:\n",confusion_matrix(yr_test,yr_pred))
    print("-" * 50)

# Print final accuracy comparison
for model, score in results_ENN.items():
    print(f"{model.__class__.__name__}: Accuracy = {score:.4f}")
```

### Comparison & Model Selection
- Random Forest performed the best with 96% accuracy, highest recall (97%), meaning it identified more churn cases correctly than the others.
- XGBoost is close with 95% accuracy and fewer false positives than Decision Tree.
- Decision Tree is slightly weaker, having more false positives and false negatives.

**Best Model = Random Forest, but XGBoost is also strong and may generalize better.**
